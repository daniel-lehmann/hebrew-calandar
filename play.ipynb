```python
import math

def f(degrees):
    
    rad = math.radians(degrees)
    e = 0.03462
    math.atan2(math.sin(rad), math.cos(rad) + e)
    # rad = math.atan2(math.sin(rad), math.cos(rad) - e) - rad
    return math.degrees(rad)

def to_dms(decimal_degrees):
    """Convert decimal degrees to (degrees, minutes, seconds)."""
    d = int(decimal_degrees)
    m_frac = abs(decimal_degrees - d) * 60
    m = int(m_frac)
    s = (m_frac - m) * 60
    return d, m, s

for deg in range(0, 181, 10):
    value = f(deg)
    d, m, s = to_dms(value)
    print(f"{deg}° -> {d}° {m}' {s:.0f}\"")
```
